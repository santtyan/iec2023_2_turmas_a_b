# ***IA de Recomendações de Filmes***

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

url = '/kaggle/input/movie-dataset/Hydra-Movie-Scrape.csv'
data = pd.read_csv(url)

data['Title'] = data['Title'].fillna('')
data['Summary'] = data['Summary'].fillna('')
```

```python
quantidade_filmes = int(input("Em dígitos, digite a quantidade de filmes que serão mostrados do dataset? "))
data.head(quantidade_filmes)
```

```python
filme = input(str("Digite o nome de um filme contido no DataSet:"))
```

```python
quantidade_filmes = int(input("Em dígitos, digite a quantidade de filmes que serão recomendados:"))
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(data['Summary'])

cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def get_top_n_recommendations(title, cosine_similarities=cosine_similarities, n = quantidade_filmes) :
    
    idx = data.index[data['Title'] == title].tolist()[0]
    
    similarity_scores = list(enumerate(cosine_similarities[idx]))
    
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    
    similarity_scores = similarity_scores[1:(n+1)]
    
    movie_indices = [x[0] for x in similarity_scores]
    
    recommended_movies = data['Title'].iloc[movie_indices].tolist()

    return recommended_movies
```

```python
movie_to_recommend = filme
recommendations = get_top_n_recommendations(movie_to_recommend)
print(f"Recomendações para '{movie_to_recommend}':")
print(recommendations)
```
